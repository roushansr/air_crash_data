## Main code to extract running through all the pages of website from 1919 - 1925 and extracting all the raw data to an excel sheet

```python
import time

# Base URL format
base_url = 'https://asn.flightsafety.org/database/year/{year}/{page}'

# Define column headers (adjust if needed)
columns = ["Date", "Aircraft", "Registration", "Operator", "Passengers", "Location", "Additional Info", "Status", "Extra"]

# Container to hold all data
all_data = []
```

```python

# Loop through years
for year in range(1919, 2026):  # 1910 to 2025 inclusive
    page = 1
    while True:
        url = base_url.format(year=year, page=page)
        print(f"Scraping Year: {year}, Page: {page} -> {url}")

        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.93 Safari/537.36'
        }
        response = requests.get(url, headers=headers)

        
        if response.status_code != 200:
            print(f" --> No page found (Status: {response.status_code}), moving to next year.")
            break  # Exit pagination loop, move to next year
        
        soup = BeautifulSoup(response.content, 'html.parser')
        
        table = soup.find('table')
        
        if not table:
            print(" --> No table found on this page, moving to next year.")
            break  # No table = no data
        
        rows = table.find_all("tr")
        
        page_data = []
        
        for row in rows:
            columns_data = row.find_all("td")
            row_data = [col.text.strip() for col in columns_data]
            if row_data:
                page_data.append(row_data)
        
        if page_data:
            print(f" --> Extracted {len(page_data)} rows.")
            all_data.extend(page_data)
        else:
            print(" --> No data rows found, moving to next year.")
            break  # No more data, stop

        page += 1  # Go to next page
        time.sleep(1)  # Be polite to server

# Now create the DataFrame
df = pd.DataFrame(all_data)

#Add headers if data length matches (watch for extra columns)
if df.shape[1] == len(columns):
    df.columns = columns

# Save to Excel
df.to_excel("air_crash_full_data.xlsx", index=False)
print("Scraping completed. Data saved to 'air_crash_full_data.xlsx'.")
```
